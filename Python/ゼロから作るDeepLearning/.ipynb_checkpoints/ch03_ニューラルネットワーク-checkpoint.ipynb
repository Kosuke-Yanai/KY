### 3.2.3 ステップ関数のグラフ

```python
import matplotlib.pyplot as plt
def step_function(x):
    return np.array(x > 0, dtype=np.int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

### 3.2.4 シグモイド関数の実装

```python
def sigmoid_function(x):
    return 1 / (1 + np.exp(-x))
x = np.array([-1.0, 1.0, 2.0])
sigmoid_function(x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid_function(x)
plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y1 = step_function(x)
y2 = sigmoid_function(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.ylim(-0.1, 1.1)
plt.show()
```

### 3.2.7 ReLu関数

```python
def relu(x):
    return np.maximum(0, x)
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x,y)
plt.ylim(-0.1, 5.0)
plt.show()
```

## 3.4 3層ニューラルネットワークの実装

```python
# 1層目の重み付き和の計算
X = np.array([1.0, 0.5])
W1 = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
B1 = np.array([0.1, 0.2, 0.3])

A1 = np.dot(X, W1) + B1
```

```python
print(A1)
```

```python
# 1層目の活性化関数(シグモイド)のプロセス
Z1 = sigmoid_function(A1)
print(Z1)
```

```python
# 1層目から2層目までのプロセス
W2 = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])
B2 = np.array([0.1, 0.2])
A2 = np.dot(Z1, W2) + B2
Z2 = sigmoid_function(A2)
print(Z2)
```

```python
# 出力層へのプロセス
def identity_function(x):
    return x

W3 = np.array([[0.1, 0.3], [0.2, 0.4]])
B3 = np.array([0.1, 0.2])
A3 = np.dot(Z2, W3) + B3
Y = identity_function(A3)
print(Y)
```

## 3.5.1 恒等関数とソフトマックス関数

```python
a = np.array([0.3, 2.9, 4.0])
exp_a = np.exp(a)
print(exp_a)
```

```python
sum_exp_a = np.sum(exp_a)
print(sum_exp_a)
```

```python
y = exp_a / sum_exp_a
print(y)
```

```python
def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    
    return y
```

```python
print(softmax(a))
```

```python
# オーバーフロー対策
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a-c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    
    return y
```

```python
print(softmax(np.array([1000, 1010, 900])))
```

```python
print(np.sum(softmax(np.array([0.3, 2.9, 4.0]))))
```
