アヤメデータに対し、混合ガウス分布を適用した場合。

```python
    from sklearn.datasets import load_iris
from sklearn.mixture import GaussianMixture

data = load_iris()

n_components = 3 #ガウス分布の数
model = GaussianMixture(n_components=n_components)
model.fit(data.data)
print(model.predict(data.data)) #クラスを予測
print(model.means_) #各ガウス分布の平均
print(model.covariances_) #各ガウス分布の分散
```

アヤメデータの描画を指定したガウス分布数によって自動描画する

```python
from sklearn.mixture import GaussianMixture
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = load_iris()

n_components = 3 #ガウス分布の数
model = GaussianMixture(n_components=n_components)
model.fit(data.data)

datas = pd.DataFrame(data.data, columns = data.feature_names) #特徴量をデータフレームへ変換
datas["class"] = model.predict(data.data) #cluster列を追加し、データに所属クラスタ情報を格納

#抽出する軸
x_axis = data.feature_names[0]
y_axis = data.feature_names[3]
    
for i in range(n_components):
    #特徴量の描画
    plt.scatter(datas[datas["class"]==i][x_axis], datas[datas["class"]==i][y_axis], label = ("class", i))

plt.xlabel(x_axis)
plt.ylabel(y_axis)

plt.legend()
plt.show()
```

さらに、等高線を追加する

```python
from sklearn.mixture import GaussianMixture
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt

data = load_iris()
N_CLASS = len(data.target_names) # 3
LEN_CLASS = len(data.target) // N_CLASS # 50

vx,vy = 0,3 # 対象の２変数
x_min, x_max = min(data.data[:,vx]), max(data.data[:,vx]) #データ列の最小、最大値を格納している
y_min, y_max = min(data.data[:,vy]), max(data.data[:,vy])

# ２変数で散布図を描画
cs = []
for c in 'rgb': #変数cにr,g,bの順に文字が格納されていく
    cs += list(c*LEN_CLASS)
plt.scatter(data.data[:,vx], data.data[:,vy], c=cs)
plt.xlabel(data.feature_names[vx])
plt.ylabel(data.feature_names[vy])
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

# 散布図に合わせた２変数で訓練
model = GaussianMixture(n_components=N_CLASS, random_state=3)
model.fit(data.data[:,[vx,vy]])

# 存在する範囲内で点群を生成
N = 100
x, y = np.meshgrid(np.linspace(x_min,x_max,N), np.linspace(y_min,y_max,N))
X = np.array([x, y]).reshape(2, -1).T
probs = model.predict_proba(X)

# 等高線図を各クラス毎に描画
cs = ['Greens', 'Blues', 'Reds'] # 散布図の色とは必ずしも対応しないことに注意
assert len(cs) == N_CLASS
for i in range(N_CLASS):
    plt.contourf(x, y, probs[:,i].reshape(N,N), cmap=cs[i], alpha=0.2)
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
plt.show()
```
