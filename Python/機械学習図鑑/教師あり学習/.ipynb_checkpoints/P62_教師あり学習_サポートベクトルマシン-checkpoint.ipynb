学習データを描画用に2つに分けて作成  
```make_blobs``` 分類データセット作成
n_samples	サンプル数  
n_features	特徴量の数  
centers	中心の個数（塊数）  
cluster_std	クラスタの標準偏差  

生成した点の座標がXに格納され、それを分類した結果のクラス数がyに入る。

````python
from sklearn.svm import LinearSVC
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X1, y1 = make_blobs(n_samples=50, n_features=2, centers=[(-1,-0.125)], cluster_std=0.3)
X2, y2 = make_blobs(n_samples=50, n_features=2, centers=[(0.5,0.5)], cluster_std=0.3)
````

2種類の学習データを描画する。  
[行:列]でスライスでき、省略した場合はすべてを指定したことになるので、[:, 0]は全ての行の0列目を取得することなる。

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(X1[:, 0], X1[:, 1], c = "orange", label = "data1")
plt.scatter(X2[:, 0], X2[:, 1], c = "blue", label = "data2")

plt.legend()
plt.show()
````

SVM解析を行うために、上記2種類のデータを結合する。  
別々にblobs生成したため、分類がどちらも"0"になっているので、y2の分類を"1"に変更する。

````python
y2[:] = 1
X_conv = np.vstack([X1, X2])
y_conv = np.hstack([y1, y2])
````

SVM解析にかける。

````python
X_train, X_test, y_train, y_test = train_test_split(X_conv, y_conv, test_size=0.3)
model = LinearSVC()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
````

データに分離直線を加えたものを描画する。

````python
plt.scatter(X1[:, 0], X1[:, 1], c = "orange", label = "data1")
plt.scatter(X2[:, 0], X2[:, 1], c = "blue", label = "data2")

x_svm = np.arange(-1, 1, 0.1)
y_svm = - model.coef_[0, 0] * x_svm + model.intercept_[0]
plt.plot(x_svm, y_svm, c = "black", label = "SVC")

plt.legend()
plt.show()
````
