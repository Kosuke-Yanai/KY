```python
import pandas as pd
item_master = pd.read_csv("../sample_100knocks/サンプルコード_20200325/1章/item_master.csv")
transaction_1 = pd.read_csv("../sample_100knocks/サンプルコード_20200325/1章/transaction_1.csv")
transaction_2 = pd.read_csv("../sample_100knocks/サンプルコード_20200325/1章/transaction_2.csv")
transaction_detail_1 = pd.read_csv("../sample_100knocks/サンプルコード_20200325/1章/transaction_detail_1.csv")
transaction_detail_2 = pd.read_csv("../sample_100knocks/サンプルコード_20200325/1章/transaction_detail_2.csv")
```

```python
transaction_detail_2.head()
```

```python
transaction = pd.concat([transaction_1, transaction_2], ignore_index=True)
print(len(transaction_1))
print(len(transaction_2))
print(len(transaction))
```

```python
transaction_detail = pd.concat([transaction_detail_1, transaction_detail_2], ignore_index=True)
print(len(transaction_detail_1))
print(len(transaction_detail_2))
print(len(transaction_detail))
```

```python
join_data = pd.merge(transaction_detail, transaction[["transaction_id", "payment_date", "customer_id"]], on="transaction_id", how="left")
join_data.head()
```

```python
join_data = pd.merge(join_data, item_master[["item_id", "item_name", "item_price"]], on="item_id", how="left")
join_data.head()
```

```python
join_data["price_sum"] = join_data["quantity"] * join_data["item_price"]
join_data.head()
```

```python
del join_data["price_sum"]
join_data.head()
```

```python
print(transaction["price"].sum())
print(join_data["price_sum"].sum())
```

```python
join_data.isnull().sum()
```

```python
join_data.describe()
```

```python
join_data["payment_date"] = pd.to_datetime(join_data["payment_date"])
join_data["payment_month"] = join_data["payment_date"].dt.strftime("%Y%m")
join_data.head()
```

```python
join_data.groupby(["payment_month", "item_name"]).sum()[["price_sum", "quantity"]]
```

```python
graph_data = pd.pivot_table(join_data, index="item_name", columns="payment_month", values=["price_sum", "quantity"], aggfunc="sum")
graph_data.head()
```

```python
graph_data = pd.pivot_table(join_data, index="payment_month", columns="item_name", values="price_sum", aggfunc="sum")
graph_data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(list(graph_data.index), graph_data["PC-A"], label="PC-A")
plt.plot(list(graph_data.index), graph_data["PC-B"], label="PC-B")
plt.plot(list(graph_data.index), graph_data["PC-C"], label="PC-C")
plt.plot(list(graph_data.index), graph_data["PC-D"], label="PC-D")
plt.plot(list(graph_data.index), graph_data["PC-E"], label="PC-E")
plt.legend
```
