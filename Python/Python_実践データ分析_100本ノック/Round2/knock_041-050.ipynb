### ノック41

```python
import pandas as pd
customer = pd.read_csv("customer_join.csv")
uselog_monthes = pd.read_csv("uselog_monthes.csv")
```

```python
customer.head()
```

```python
uselog_monthes.head()
```

```python
year_monthes = list(uselog_monthes["年月"].unique())
year_monthes
```

```python
uselog = pd.DataFrame()
for i in range(1, len(year_monthes)):
    tmp = uselog_monthes.loc[uselog_monthes["年月"]==year_monthes[i]]
    tmp.rename(columns={"count":"count_0"}, inplace=True)
    tmp_before = uselog_monthes.loc[uselog_monthes["年月"]==year_monthes[i-1]]
    
    del tmp_before["年月"]
    tmp_before.rename(columns={"count":"count_1"}, inplace=True)
    tmp = pd.merge(tmp, tmp_before, on="customer_id", how="left")
    uselog = pd.concat([uselog, tmp], ignore_index=True)
uselog.head()
```

```python
uselog["年月"]
```

### ノック42

```python
from dateutil.relativedelta import relativedelta
exit_customer = customer.loc[customer["is_deleted"]==1]
exit_customer.head()
```

```python
exit_customer["exit_date"] = None
exit_customer["end_date"] = pd.to_datetime(exit_customer["end_date"])
for i in range(len(exit_customer)):
   exit_customer["exit_date"].iloc[i] = exit_customer["end_date"].iloc[i] - relativedelta(months=1)
exit_customer["年月"] =  exit_customer["exit_date"].dt.strftime("%Y%m")
exit_customer.head()
```

```python
uselog["年月"] = uselog["年月"].astype(str)
uselog.head()
```

```python
exit_uselog = pd.merge(uselog, exit_customer, on=["customer_id", "年月"], how="left")
exit_uselog.head()
```

```python
print(len(exit_uselog))
```

```python
exit_uselog = exit_uselog.dropna(subset=["name"])
exit_uselog.head()
```

```python
print(len(exit_uselog))
```

```python
len(exit_uselog["customer_id"].unique())
```

### ノック43

```python
conti_customer = customer.loc[customer["is_deleted"]==0]
conti_customer.head()
```

```python
conti_uselog = pd.merge(uselog, conti_customer, on=["customer_id"], how="left")
conti_uselog.head()
```

```python
len(conti_uselog)
```

```python
conti_uselog = conti_uselog.dropna(subset=["name"])
len(conti_uselog)
```

```python
conti_uselog = conti_uselog.sample(frac=1).reset_index(drop=True)
conti_uselog = conti_uselog.drop_duplicates(subset="customer_id")
print(len(conti_uselog))
conti_uselog.head()
```

```python
predict_data = pd.concat([conti_uselog, exit_uselog], ignore_index=True)
print(len(predict_data))
predict_data.head()
```

### ノック44

```python
predict_data["period"] = 0
predict_data["now_date"] = pd.to_datetime(predict_data["年月"], format="%Y%m")
predict_data.head()
```

```python
predict_data["start_date"] = pd.to_datetime(predict_data["start_date"])
for i in range(len(predict_data)):
    delta = relativedelta(predict_data["now_date"].iloc[i], predict_data["start_date"].iloc[i])
    predict_data["period"].iloc[i] = int(delta.years*12 + delta.months)
predict_data.head()
```

### ノック45

```python
predict_data.isna().sum()
```

```python
predict_data = predict_data.dropna(subset=["count_1"])
predict_data.isna().sum()
```

### ノック46

```python
target_col = ["campaign_name", "class_name", "gender", "count_1", "routine_flg", "period", "is_deleted"]
predict_data = predict_data[target_col]
predict_data.head()
```

```python
predict_data = pd.get_dummies(predict_data)
predict_data.head()
```

```python
del predict_data["campaign_name_通常"]
del predict_data["class_name_デイタイム"]
del predict_data["gender_M"]
predict_data.head()
```

### ノック47

```python
from sklearn.tree import DecisionTreeClassifier
import sklearn.model_selection

exit = predict_data.loc[predict_data["is_deleted"]==1]
conti = predict_data.loc[predict_data["is_deleted"]==0].sample(len(exit))
conti.head()
```

```python
X = pd.concat([exit, conti], ignore_index=True)
y = X["is_deleted"]
del X["is_deleted"]
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,y)

model = DecisionTreeClassifier(random_state=0)
model.fit(X_train, y_train)
y_test_pred = model.predict(X_test)
print(y_test_pred)
```

```python
result_test = pd.DataFrame({"y_test":y_test, "y_pred":y_test_pred})
result_test.head()
```

### ノック48

```python
correct = len(result_test.loc[result_test["y_test"]==result_test["y_pred"]])
data_count = len(result_test)
score_test = correct / data_count
score_test
```

```python
print(model.score(X_test, y_test))
print(model.score(X_train, y_train))
```

```python
X = pd.concat([exit, conti], ignore_index=True)
y = X["is_deleted"]
del X["is_deleted"]
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,y)

model = DecisionTreeClassifier(random_state=0, max_depth=5)
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
print(model.score(X_train, y_train))
```

```python
count_1 = 2
routine_flg = 1
period = 7
campaign_name = "通常"
class_name = "ナイト"
gender = "M"

if campaign_name == "入会費半額":
    campaign_name_list = [1,0]
elif campaign_name == "入会費無料":
    campaign_name_list = [0,1]
elif campaign_name == "通常":
    campaign_name_list = [0,0]

if class_name == "オールタイム":
    class_name_list = [1,0]
if class_name == "デイタイム":
    class_name_list = [0,1]
if class_name == "ナイト":
    class_name_list = [0,0]

if gender == "F":
    gender_list = [1]
if gender == "M":
    gender_list = [0]
 
imput_data = [count_1, routine_flg, period]
imput_data.extend(campaign_name_list)
imput_data.extend(class_name_list)
imput_data.extend(gender_list)
```

```python
print(model.predict([imput_data]))
print(model.predict_proba([imput_data]))
```
