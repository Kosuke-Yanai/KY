### ノック1

```python
import pandas as pd
customer_master = pd.read_csv("../sample_100knocks/サンプルコード_20200325/1章/customer_master.csv")
customer_master.head()
```

```python
item_master = pd.read_csv("../sample_100knocks/サンプルコード_20200325/1章/item_master.csv")
item_master.head()
```

```python
transaction_1 = pd.read_csv("../sample_100knocks/サンプルコード_20200325/1章/transaction_1.csv")
transaction_1.head()
```

```python
transaction_2 = pd.read_csv("../sample_100knocks/サンプルコード_20200325/1章/transaction_2.csv")
transaction_2.head()
```

```python
transaction_detail_1 = pd.read_csv("../sample_100knocks/サンプルコード_20200325/1章/transaction_detail_1.csv")
transaction_detail_1.head()
```

```python
transaction_detail_2 = pd.read_csv("../sample_100knocks/サンプルコード_20200325/1章/transaction_detail_2.csv")
transaction_detail_2.head()
```

### ノック2

```python
transaction = pd.concat([transaction_1, transaction_2], ignore_index=True)
transaction.head()
```

```python
print("transaction_1:", len(transaction_1))
print("transaction_2:", len(transaction_2))
print("transaction_sum:", len(transaction))
```

```python
transaction_detail = pd.concat([transaction_detail_1, transaction_detail_2], ignore_index=True)
transaction_detail.head()
```

```python
print("transaction_detail_1:", len(transaction_detail_1))
print("transaction_detail_2:", len(transaction_detail_2))
print("transaction_detail_sum:", len(transaction_detail))
```

### ノック3

```python
join_data = pd.merge(transaction_detail, transaction[["transaction_id", "payment_date", "customer_id"]], on="transaction_id", how="left")
join_data.head()
```

```python
print(len(transaction_detail))
print(len(transaction))
print(len(join_data))
```

### ノック4

```python
join_data = pd.merge(join_data, customer_master, on = "customer_id", how="left")
join_data = pd.merge(join_data, item_master, on = "item_id", how="left")
join_data.head()
```

### ノック5

```python
join_data["price"] = join_data["quantity"] * join_data["item_price"]
join_data.head()
```

### ノック6

```python
print(join_data["price"].sum())
print(transaction["price"].sum())
join_data["price"].sum() == transaction["price"].sum()
```

### ノック8

```python
join_data.dtypes
```

```python
join_data["payment_date"].head()
```

```python
join_data["payment_date"] = pd.to_datetime(join_data["payment_date"])
join_data["payment_month"] = join_data["payment_date"].dt.strftime("%Y%m")
join_data[["payment_date", "payment_month"]].head()
```

```python
join_data.groupby("payment_month").sum()["price"]
```

### ノック9

```python
join_data.groupby(["payment_month", "item_name"]).sum()[["price", "quantity"]]
```

```python
pd.pivot_table(join_data, index="item_name", columns="payment_month", values=["price", "quantity"], aggfunc="sum")
```

### ノック10

```python
graph_data = pd.pivot_table(join_data, index="payment_month", columns="item_name", values="price", aggfunc="sum")
graph_data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(list(graph_data.index), graph_data["price"]["PC-A"], label="PC-A")
plt.legend()
```
