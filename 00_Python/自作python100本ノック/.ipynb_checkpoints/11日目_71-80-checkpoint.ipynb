## Q71: 連続した数かどうか

```
問題：1118 のような、3 つ以上の同じ数字が連続して並んだ 4 桁の整数を 良い整数 とします。 4 桁の整数 N が与えられるので、N が 良い整数 かどうかを答えてください。
```

````python
import random

def NumChekcer(num):
    n = str(num)
    print('Number:', num)
    if n[0] == n[1] == n[2] or n[1] == n[2] == n[3]:
        print("Yes, it is a good number")
    else:
        print("No, it isn't good nomber")
    

num = random.randrange(1000, 9999)
NumChekcer(num)
````

## Q72:　リュカ数

```
問題： 整数 N が与えられるので、N 番目のリュカ数を求めてください。 ただし、リュカ数は i 番目のリュカ数を Li とすると、 L0=2 L1=1 Li=Li−1+Li−2(i≧2) と定義される数とします。
```

````python
def LyucaNum(num):
    if num == 0:
        return 2
    elif num == 1:
        return 1
    else:
        L =  LyucaNum(num - 1) + LyucaNum(num - 2)
        return L

LyucaNum(4)
````

## Q73: 途中式の自動作成

```
問題：
4 つの 0 以上 9 以下の整数 A,B,C,D を順に受けとります。
A op1 B op2 C op3 D = 7 となるように、op1,op2,op3 に + か - を入れて式を自動で作り表示しなさい。 なお、答えが存在しない場合は"impossible"と表示し、また答えが複数存在する場合は全て表示させなさい。
```

````python
def AnswerSeven(A, B, C, D):
    answer1 = 0
    answer2 = 0
    answer3 = 0
    judge = 0
    for i in range(0,2):
        if i % 2 == 0:
            answer1 = A + B
            op1 = '+'
        else:
            answer1 = A - B
            op1 = '-'
        for j in range(0,2):
            if j % 2 == 0:
                answer2 = answer1 + C
                op2 = '+'
            else:
                answer2 = answer1 - C
                op2 = '-'
            for k in range(0,2):
                if k % 2 == 0:
                    answer3 = answer2 + D
                    op3 = '+'
                else:
                    answer3 = answer2 - D
                    op3 = '-'
                if answer3 == 7:
                    judge = 1
                    print('A', op1, 'B', op2, 'C', op3, 'D = ', answer3)
    if judge == 0:
        print('impossible')
AnswerSeven(7, 1, 1, 0)
````

````python
# 回答例。'+-'の文字列を定義するとこがスマート(eval)
def put_formula(n):
    a,b,c,d = list(str(n))
    sign = "+-"
    aws_list = []
    for i in range(2):  # 1つ目の記号
        for j in range(2):  # 2つ目の記号
            for k in range(2):  # 3つ目の記号
                if eval(a+sign[i]+b+sign[j]+c+sign[k]+d) == 7:
                    aws = (str(a+sign[i]+b+sign[j]+c+sign[k]+d)+"=7")
                    aws_list.append(aws)
                    print(aws)
    if len(aws_list) == 0:
        print("impossible")

put_formula(1161)
````

## Q75 :文章の最初の文字を表示

```
問題：　与えられた文章の内、最初の文字を表示しなさい
条件：　'.'や','や空白は文字としてカウントしない。
例：　
first_word("Hello world") == "Hello"
first_word(" a word ") == "a"
first_word("greetings, friends") == "greetings"
first_word("... and so on ...") == "and"
first_word("Hello.World") == "Hello"
```

````python
import re

def first_word(text):
    word = re.findall(r'\w+', text)
    print(word[0])

first_word("Hello world")
first_word(" a word ")
first_word("greetings, friends")
first_word("... and so on ...")
first_word("Hello.World")
````

## Q76 :テキスト内で２回目にシンボルが出てくるのは何番目か

```
問題：テキスト文と文字がそれぞれ与えられる。
　　　与えられたテキスト文のうち、文字が２回目に出てくるのはテキスト文のうち何番目かを表示せよ。
```

````python
def symbol_checker(text, symbol):
    symbol_order_1st = text.find(symbol)
    symbol_order_2nd = text.find(symbol, symbol_order_1st + 1) 
    if symbol_order_1st >= 0:
        if symbol_order_2nd >= 0:
            print('symbol:', symbol, ' order:', symbol_order_2nd)
        else:
            print('this symbol appears only once!')
    else:
        print('no such a symbol in text!')
        
text = 'if i was a chid, i want to play outside.'
symbol = 'w'
symbol_checker(text, symbol)
````

````python
# 回答例。カウント値が2に達したら即終了。シンプル。
def second_index(text, symbol):
    count = 0
    for i in range(len(text)):
        if text[i] == symbol:
            count += 1
            if count == 2:
                return i
    return None

second_index("I am a good student but you are not a good student", "g")
````

## Q78:単語の出現頻度を表示せよ

```
問題：文章と、いくつかの単語が与えられる。
文章のうち、それぞれの単語が何回含まれているか表示せよ。
例：popular_words('''When I was OneI had just begunWhen I was TwoI was nearly new
''', ['i', 'was', 'three', 'near']) == {'i': 4, 'near': 0, 'three': 0, 'was': 3}
```

````python
def WordCounter(text, words):
    counter = {}
    text = text.split()
    for word in words:
        counter[word] = text.count(word)
    return counter

text = 'When I was OneI had just begunWhen I was TwoI was nearly new'
words = ['I', 'was', 'three', 'near']

WordCounter(text, words)
````
