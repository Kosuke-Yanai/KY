#### ノック41：データを読み込んで利用データを整形しよう

```python
import pandas as pd
uselog_month = pd.read_csv("uselog_month.csv")
uselog_month.head()
```

```python
customer = pd.read_csv("customer_join.csv")
customer.head()
```

```python
year_month = list(uselog_month["年月"].unique())
uselog = pd.DataFrame()
for i in range(1, len(year_month)):
    tmp = uselog_month.loc[uselog_month["年月"] == year_month[i]]
    tmp.rename(columns={"count":"count_0"}, inplace=True)
    tmp_before =  uselog_month.loc[uselog_month["年月"] == year_month[i-1]]
    del tmp_before["年月"]
    tmp_before.rename(columns={"count":"count_1"}, inplace=True)
    tmp = pd.merge(tmp, tmp_before, on="customer_id", how="left")
    uselog = pd.concat([uselog, tmp], ignore_index=True)
uselog.head()
```

#### ノック42：退会前月の退会顧客データを作成しよう

```python
from dateutil.relativedelta import relativedelta
exit_customer = customer.loc[customer["is_deleted"]==1]
exit_customer["exit_date"] = None
exit_customer["end_date"] = pd.to_datetime(exit_customer["end_date"])
for i in range(len(exit_customer)):
    exit_customer["exit_date"].iloc[i] = exit_customer["end_date"].iloc[i] - relativedelta(months=1)
exit_customer["年月"] = exit_customer["exit_date"].dt.strftime("%Y%m")
uselog["年月"] = uselog["年月"].astype(str)
exit_uselog = pd.merge(uselog, exit_customer, on=["customer_id", "年月"], how="left")
print(len(exit_uselog))
exit_uselog.head()                                                                          
```

```python
exit_uselog = exit_uselog.dropna(subset=["name"])
```

```python
exit_uselog.head()
```

```python
print(len(exit_uselog))
print(len(exit_uselog["customer_id"].unique()))
```

#### ノック43：継続顧客のデータを作成しよう

```python
conti_customer = pd.merge(uselog, customer.loc[customer["is_deleted"]==0], on="customer_id", how="left")
```

```python
print(len(conti_customer))
conti_customer = conti_customer.dropna(subset=["name"])
print(len(conti_customer))
```

```python
conti_customer = conti_customer.sample(frac=1).reset_index(drop=True)
conti_customer = conti_customer.drop_duplicates(["customer_id"])
print(len(conti_customer))
conti_customer.head()
```

```python
predict_data = pd.concat([conti_customer, exit_uselog], ignore_index=True)
predict_data.head()
```

```python
print(len(predict_data))
```

#### ノック44：予測する月の在籍期間を作成しよう

```python
predict_data["period"] = 0
for i in range(len(predict_data)):
    delta = relativedelta(pd.to_datetime(predict_data["年月"].iloc[i], format="%Y%m"), pd.to_datetime(predict_data["start_date"].iloc[i]))
    predict_data["period"].iloc[i] = delta.years*12 + delta.months
predict_data.head()
```

#### ノック45：欠損値を除去しよう

```python
predict_data.isnull().sum()
```

```python
predict_data.dropna(subset=["count_1"], inplace=True)
```

```python
predict_data.isnull().sum()
```

#### ノック46：文字列型の変数を処理できるように整形しよう

```python
predict_data.head()
```

```python
target_col = ["campaign_name", "class_name", "gender", "count_1", "routine_flg", "period", "is_deleted"]
predict_data  = predict_data[target_col]
predict_data.head()
```

```python
predict_data = pd.get_dummies(predict_data)
predict_data.head()
```

```python
del predict_data["gender_M"]
```

```python
del predict_data["campaign_name_通常"]
del predict_data["class_name_ナイト"]
```

#### ノック47：決定木を用いて退会予測モデルを作成してみよう

```python
exit = predict_data.loc[predict_data["is_deleted"]==1]
conti = predict_data.loc[predict_data["is_deleted"]==0]
print(len(exit))
print(len(conti))
```

```python
conti = conti.sample(n=len(exit)).reset_index(drop=True)
print(len(conti))
```

```python
from sklearn.tree import DecisionTreeClassifier
import sklearn.model_selection

X = pd.concat([exit, conti], ignore_index=True)
y = X["is_deleted"]
del X["is_deleted"]

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,y)
model = DecisionTreeClassifier(random_state=0)
model.fit(X_train, y_train)
y_test_pred = model.predict(X_test)
print(y_test_pred)
```

```python
results_test = pd.DataFrame({"y_test":y_test, "y_pred":y_test_pred})
results_test.head()
```

#### ノック48：予測モデルの評価を行い、モデルのチューニングをしてみよう

```python
match_num = len(results_test.loc[results_test["y_test"]==results_test["y_pred"]])
print(match_num / len(results_test))
```

```python
print(model.score(X_test, y_test))
print(model.score(X_train, y_train))
```

```python
X = pd.concat([exit, conti], ignore_index=True)
y = X["is_deleted"]
del X["is_deleted"]

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,y)
model = DecisionTreeClassifier(random_state=0, max_depth=3)
model.fit(X_train, y_train)
y_test_pred = model.predict(X_test)

print(model.score(X_test, y_test))
print(model.score(X_train, y_train))
```

#### ノック49：モデルに寄与している変数を確認しよう

```python
importance = pd.DataFrame({"feature_names":X.columns, "coefficient":model.feature_importances_})
importance
```

#### ノック50：顧客の退会を予測しよう

```python
count_1 = 5
routine_flg = 1
period = 10
campaign_name = "入会費無料"
class_name = "ナイト"
gender = "M"

if campaign_name == "入会費半額":
    campaign_name_list = [1, 0]
elif campaign_name == "入会費無料":
    campaign_name_list[:2] = [0, 1]
elif campaign_name == "通常":
    campaign_name_list = [0, 0]

if class_name == "オールタイム":
    class_name_list = [1, 0]
elif class_name == "デイタイム":
    class_name_list = [0, 1]
elif class_name == "ナイト":
    class_name_list = [0, 0]

if gender == "F":
    gender_list = [1]
elif gender == "M":
    gender_list = [0]

input_data = [count_1, routine_flg, period]
input_data.extend(campaign_name_list)
input_data.extend(class_name_list)
input_data.extend(gender_list)

print(model.predict([input_data]))
print(model.predict_proba([input_data]))
```
