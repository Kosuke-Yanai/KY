## 4.3.1 微分

```python
def numerical_diff(f, x):
    h = 10e-4
    return (f(x+h) - f(x-h)) / (2*h)
```

## 4.3.2 数値微分の例

```python
def function_1(x):
    return 0.01*x**2 + 0.1*x
```

```python
import numpy as np
import matplotlib.pylab as plt

x = np.arange(0.0, 20.0, 0.1)
y = function_1(x)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y)
plt.show()
```

```python
numerical_diff(function_1, 10)
```

```python
def tangent_line(f, x):
    d = numerical_diff(f, x)
    print(d)
    y = f(x) - d*x
    return lambda t: d*t + y
```

```python
x = np.arange(0.0, 20.0, 0.1)
y = function_1(x)

tf = tangent_line(function_1, 5)
y2 = tf(x)

plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```
