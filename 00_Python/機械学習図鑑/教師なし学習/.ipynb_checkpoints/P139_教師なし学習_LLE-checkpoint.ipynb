SwissRollのデータセットを使用してLLEに適用するサンプルコード

```python
from sklearn.datasets import samples_generator
from sklearn.manifold import LocallyLinearEmbedding

data, color = samples_generator.make_swiss_roll(n_samples=1500)

n_neighbors = 15 #近傍点の数
n_components = 2 #削減後の次元数
model = LocallyLinearEmbedding(n_neighbors=n_neighbors, n_components=n_components)

model.fit(data)
print(model.transform(data))
```

解析したLLEの結果をプロットして可視化する

```python
from sklearn.mixture import GaussianMixture
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(model.transform(data)[:, 0], model.transform(data)[:, 1], c = color)
plt.show()
```
