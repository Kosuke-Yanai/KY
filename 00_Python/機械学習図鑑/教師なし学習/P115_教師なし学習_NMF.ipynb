NMF実行サンプルコード

```python
from sklearn.decomposition import NMF
from sklearn.datasets.samples_generator import make_blobs

centers = [[5, 10, 5], [10, 4, 10], [6, 8, 8]]
V, _ = make_blobs(centers=centers)

n_components = 3
model = NMF(n_components=n_components)
model.fit(V)

W = model.transform(V) #分解後の行列
H = model.components_
```

データセットの描画  
3次元グラフの作成順序  
 1. figureで2次元の図を生成する  
 2. その後，Axes3D関数で3次元にする   
 (奥の点ほど、色が薄く表示される)

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure()
ax = Axes3D(fig)

V_x_0 = V[_==0,0:1]
V_y_0 = V[_==0,1:2]
V_z_0 = V[_==0,2:3]

V_x_1 = V[_==1,0:1]
V_y_1 = V[_==1,1:2]
V_z_1 = V[_==1,2:3]

V_x_2 = V[_==2,0:1]
V_y_2 = V[_==2,1:2]
V_z_2 = V[_==2,2:3]

V_nmf = np.dot(W, H)
V_nmf_x = V_nmf[:,0]
V_nmf_y = V_nmf[:,1]
V_nmf_z = V_nmf[:,1]

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")

ax.scatter(V_x_0, V_y_0, V_z_0, c = "blue", label = "class0")
ax.scatter(V_x_1, V_y_1, V_z_1, c = "orange", label = "class1")
ax.scatter(V_x_2, V_y_2, V_z_2, c = "green", label = "class2")
ax.scatter(V_nmf_x, V_nmf_y, V_nmf_z, c = "black", label = "NMF")
plt.show()
```
